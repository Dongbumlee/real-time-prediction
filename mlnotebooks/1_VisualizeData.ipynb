# Visualize the training Dataset

```python
from azureml.core import Workspace
from azureml.core.model import Model
from azureml.core.dataset import Dataset

ws = Workspace.from_config()
trainingData = ws.datasets['qualitydataset'].to_pandas_dataframe()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

features = ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9', 'S10']

fig_size = plt.rcParams["figure.figsize"] 
print ("Current size:", fig_size)
fig_size[0] = 15
fig_size[1] = 15
plt.rcParams["figure.figsize"] = fig_size

trainingData[features].hist()
```

```python
# Plotting correlation matrix
import matplotlib.pyplot as plt

df = trainingData[features]

#corr = df.corr()
#corr.style.background_gradient(cmap='coolwarm')

f = plt.figure(figsize=(10, 10))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.shape[1]), df.columns, fontsize=11, rotation=45)
plt.yticks(range(df.shape[1]), df.columns, fontsize=11)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=11)
plt.title('Correlation Matrix', fontsize=16);
```
